```python
import numpy as np
import matplotlib.pyplot as mt
import pandas as pd
```

```python
data = pd.read_csv("C:\\Users\\pulla\\Downloads\\Store_data.csv")
data.head()
```

```python
transactions=[]
for i in range(0, 1000):
    transactions.append([str(data.values[i,j]) for j in range(0,20)])
```

```python
from mlxtend.frequent_patterns import apriori
import apyori
```

```python
rules = apyori.apriori(transactions = transactions, min_support=0.003, min_confidence = 0.2, min_lift=3, min_length=2, max_length=2)

```

```python
associations = list(rules)
associations
```
